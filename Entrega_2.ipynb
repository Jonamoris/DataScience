```python

#1. Convertimos el campo Date en fecha y Agregamos una columna MES

definitivo["Fecha"] =pd.to_datetime(definitivo["date"]) 
definitivo["Mes"] = definitivo["Fecha"].dt.month


#2. Construimos un indice por mes y Por valor

MesIndice = definitivo["Mes"].value_counts() #Contamos por cada mes la cantidad de registros que se repiten (Cantidad de vuelos)
Vuelo = definitivo["flight_number"].count()
```

```python
#Armamos el nuevo DATAFRAME CON VALORES por Mes / Cambiamos la columna por el nombre "cantidad vuelos"


dfCantidadVuelos = pd.DataFrame(MesIndice)
dfCantidadVuelos.rename(columns={'Mes':'CantidadVuelos',
                        },
               index={0:'cero',1:'uno'},
               inplace=True)

```

```python
#GRAFICO 1

# La función sns.lmplot ajustará una línea a sus datos

sns.lmplot(x='scheduled_elapsed_time', y='arrival_delay', data=definitivo).set(title="Correlacion entre el tiempos Estimado y demoras de vuelos");

#Concluimos que la mayoria de atrasos se dan en los vuelos de corta duracion. A mayor duracion de vuelo, menor probabilidad de demora
```

```python
#GRAFICO 2


sns.distplot(definitivo['arrival_delay']).set(title="Distribucion Tiempo Demora Vuelos");

#En este grafico, se observa que la mayor parte de los vuelos no poseen demora.
```

```python
#GRAFICO 3- Puntos segun distribucion por visibilidad


Puntos=sns.scatterplot(data=definitivo, x="HourlyVisibility_x", y="flight_number");
Puntos.set(title = "Distribucion vuelos segun Visibilidad Nocturna");

#Se observa que la mayoria de los vuelos, tienen una salida con una visibilidad buena.
```
